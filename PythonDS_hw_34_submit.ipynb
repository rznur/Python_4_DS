```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
# SVG (Scalable Vector Graphics) - масштабируемая векторная графика
%config InlineBackend.figure_format = 'svg'

x = [1,   2,   3,   4,   5, 6,   7]
y = [3.5, 3.8, 4.2, 4.5, 5, 5.5, 7]

# Линейная диаграмма (график)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(3,8) 
plt.xlim(0,8)
plt.grid(True)
plt.show()
```
