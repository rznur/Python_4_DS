```python
check_test = pd.DataFrame({
    "y_valid": y_valid,
    "y_pred": y_pred.flatten(),
})

check_test.head(10)
```

```python
from sklearn.metrics import r2_score

r2_score_full = r2_score(y_valid, y_pred)

r2_score_full
#0.7806334479233576
```

```python
#checking for overfitting
y_pred_train = lr.predict(X_train)


r2_score_QC_full = r2_score(y_train, y_pred_train)
r2_score_QC_full
#0.7894598957982276
```

####  Function for merging

```python
def join_stats1(df,stats1,source_df, gr_bin, mean_price, mode='train'):
    df=pd.merge(df,stats1,on=['DistrictId',gr_bin],how='left')
    if mode=='test':
        df[mean_price]=df[mean_price].fillna(source_df[mean_price].mean())
    return df
```
